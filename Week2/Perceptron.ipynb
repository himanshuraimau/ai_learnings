```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Define the positive and negative classes
positive_points = np.array([(-2,3),(0,1),(2,-1)])
negative_points=np.array([(-2,1),(0,-1),(2,-3)])
```

```python
#plot the  initial data points
plt.figure(figsize=(8,6))
plt.scatter(positive_points[:,0],positive_points[:,1],color='blue',marker='o',label='Positive Class')
plt.scatter(negative_points[:,0],negative_points[:,1],color='red',marker='x',label='Negative Class')
plt.legend(loc='best')
plt.title('Initial Data Points')
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid(True)
plt.show()
```

```python
# Combine all data points
X = np.vstack((positive_points,negative_points))
# Define the target values
Y = np.array([1]*len(positive_points) + [-1]*len(negative_points))
```

```python
# Define the perceptron algorithm
def perceptron(X,y,steps=10):
    theta = np.zeros(X.shape[1])
    theta_0 = 0
    for t in range(steps):
        for i,x in enumerate(X):
            if y[i]*(np.dot(theta,x)+theta_0) <= 0:
                theta += y[i]*x
                theta_0 += y[i]
                break
        # Plot after each iteration
        plt_decision_boundary(X,y,theta,t+1)
    return theta,theta_0
```

```python
def plt_decision_boundary(X, y, theta, iteration):
    plt.figure(figsize=(8,6))
    plt.scatter(positive_points[:,0], positive_points[:,1], color='blue', marker='o', label='Positive Class')
    plt.scatter(negative_points[:,0], negative_points[:,1], color='red', marker='x', label='Negative Class')
    
    # Plot decision boundary
    x_values = np.linspace(-3, 3, 2)
    
    # Check for theta[1] != 0
    if theta[1] != 0:
        y_values = (-theta_0 - theta[0] * x_values) / theta[1]
        plt.plot(x_values, y_values, 'g')
    else:
        plt.axvline(x=-theta_0)
    
    plt.xlim(-3, 3)
    plt.ylim(-4, 4)
    plt.legend(loc='best')
    plt.title(f'Decision Boundary at iteration {iteration}')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.grid(True)
    plt.show()
    
```

```python
theta,theta_0 = perceptron(X,Y,steps=10)
```
